### Outlier Checking

```python
Q1 = df['Score_num'].quantile(0.25)
Q3 = df['Score_num'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Score_num'] < lower_bound) | (df['Score_num'] > upper_bound)]

outliers
```

#### This says there are no specific outlliers , everything is in the range

### Box Plot

```python
import matplotlib.pyplot as plt

df.boxplot(column='Score_num', by='Batch')
plt.title('Score Distribution by Batch')
plt.suptitle('')
plt.xlabel('Batch')
plt.ylabel('Score')
plt.show()
```

### Voilin Plot

```python
plt.figure()
sns.violinplot(x='Batch', y='Score_num', data=df)
plt.title('Score Density by Batch')
plt.xlabel('Batch')
plt.ylabel('Score')
plt.show()
```

####  AI_ELITE_7	Most students score between 5 and 6
#### AI_ELITE_6	Students are spread across 3 to 5 with some very low and very high
#### AI_ELITE_4	Most students are around 3 to 4

### Bar Graph

```python
plt.figure()
sns.barplot(x='Batch', y='Score_num', data=df)
plt.title('Average Score by Batch')
plt.xlabel('Batch')
plt.ylabel('Average Score')
plt.show()
```

#### The highest average is AI_ELITE_7

### Key Findings

####  Which batch is performing the best?
- AI_ELITE_7 students score the highest on average and their performance is also the most consistent.

####  How are students performing overall?
- Most students score between 3 and 5 out of 7, which shows average performance for most learners

####  Are there students who need urgent support?
- Yes. Some students scored 0 or 1, which shows they are at risk and need immediate attention.

#### Where are the weakest students mostly found?
- Most of the very low scores (0–1) appear in AI_ELITE_4 and AI_ELITE_6.

#### How big is the difference between batches?
- The average score difference between AI_ELITE_7 and AI_ELITE_4 is more than 1.25 points on a 7-point scale, which is a large learning gap.

#### How can the organization improve results?
- By providing the teaching style of AI_ELITE_7, supporting weak students in AI_ELITE_4, and mentoring AI_ELITE_6, overall performance can improve a lot.
